# Notebook 02: PPC Analysis

Compute and visualize Performance Portability Coefficients across all configurations.

**Prerequisite:** Run `python analysis/compute_ppc.py` first.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ppc = pd.read_csv('../data/processed/ppc_results.csv')
print(ppc.shape)
ppc.groupby(['abstraction', 'ppc_tier'])['ppc'].agg(['mean', 'min', 'max']).round(3)
```

```python
# Heatmap: PPC by abstraction × kernel
pivot = ppc.pivot_table(values='ppc', index='abstraction', columns='kernel', aggfunc='mean')
fig, ax = plt.subplots(figsize=(12, 4))
sns.heatmap(pivot, ax=ax, annot=True, fmt='.2f', cmap='RdYlGn', vmin=0, vmax=1)
ax.set_title('PPC Summary: Abstraction × Kernel')
plt.tight_layout()
plt.savefig('../paper/figures/fig17_ppc_summary.pdf', bbox_inches='tight')
```
