# Notebook 03: Overhead Attribution

Decompose abstraction overhead into: kernel launch, synchronization, host framework, memory transfer, compiler quality.

**Prerequisite:** Run `python analysis/overhead_attribution.py` first.

```python
import pandas as pd
import matplotlib.pyplot as plt

oh = pd.read_csv('../data/processed/overhead_breakdown.csv')
print(f"Flagged configurations: {len(oh)}")

# Dominant category distribution
oh['dominant_category'].value_counts().plot.bar()
plt.title('Dominant Overhead Category Distribution')
plt.tight_layout()
```

```python
# Stacked bar: overhead components per abstraction
components = ['kernel_launch_ms', 'host_framework_ms', 'memory_transfer_ms', 'compiler_quality_ms']
by_abs = oh.groupby('abstraction')[components].mean()
by_abs.plot.bar(stacked=True, figsize=(10, 5))
plt.title('Average Overhead Components by Abstraction')
plt.ylabel('ms')
plt.tight_layout()
plt.savefig('../paper/figures/fig18_overhead_breakdown.pdf', bbox_inches='tight')
```
